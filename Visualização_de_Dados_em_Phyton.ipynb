```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid") #opes: darkgrid, whitegrid, dark, white, ticks

df_tips = sns.load_dataset('tips')
print(df_tips)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])
#media por sexo

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum)
#O motivo para optar pelo "barplot()" muitas vezes se baseia nos parmetros adicionais e na flexibilidade que ele oferece.
#Vamos dar destaque ao parmetro "estimator," que, por padro, calcula a mdia.

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len)
#A funo "barplot()" do Seaborn apresenta uma variedade de opes estTSTICAS,
#permitindo que os cientistas de dados escolham a mtrica que melhor se ajusta aos seus objetivos.
#Por exemplo, voc pode calcular a soma, a contagem ou at mesmo outras mtricas personalizadas.
#Isso  particulamente til quando voc deseja exibir informaes diferentes nas barras,
#como a quantidade (len) ou a soma (sum) dos valores, em vez da mdia.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

df = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df, estimator=sum, ci=None, palette="Set2")
plt.xlabel('Perodo (Time)')
plt.ylabel('Total de Gastos')
plt.title('Total de Gastos por Perodo (Almoo ou Jantar)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df) #estimator=sum, ci=None, palette="Set2"
plt.xlabel('Perodo (Time)')
plt.ylabel('Mdia de Gastos')
plt.title('Mdia de Gastos por Perodo (Almoo ou Jantar)')
plt.show()
```

```python
#Crie um grfico de barras com Seaborn para mostrar a mdia da gorjeta por perodo
plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df, palette="Set3")
plt.xlabel('Perodo (Time)')
plt.ylabel('Mdia de Gojeta')
plt.title('Mdia da Gorjeta por Perodo (Almoo ou Jantar)')
plt.show()
```
